```python
df['Year'] = df['Year'].fillna(df['Year'].mean())
df = df.dropna(axis=0)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
data = df2.copy()
encoder = LabelEncoder()

feature = ["Platform", "Genre","Publisher","NA_Sales"]


for col in feature:
    data[col] = encoder.fit_transform(df2[col])
```

### Modeling

```python
X = data[['Platform', 'Genre','Publisher',]].values
y = data[['Global_Sales']].values
```

```python
from sklearn.model_selection import train_test_split
from sklearn import linear_model
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state = 42)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import Ridge, Lasso, ElasticNet, LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
```

```python
pipelines = {
    'rf':make_pipeline(RandomForestRegressor(random_state=42)),
    'gb':make_pipeline(GradientBoostingRegressor(random_state=42)),
    'ridge':make_pipeline(Ridge(random_state=42)),
    'lasso':make_pipeline(Lasso(random_state=42)),
    'enet':make_pipeline(ElasticNet(random_state=42)),
    'reg':make_pipeline(LinearRegression())
}
```

```python
hypergrid = {
    'rf':{
        'randomforestregressor__min_samples_split':[2,4,6],
        'randomforestregressor__min_samples_leaf':[1,2,3]
    },
    'gb':{
        'gradientboostingregressor__alpha':[0.001,0.005,0.01,0.05,0.1,0.5,0.99]
    },
    'ridge':{
        'ridge__alpha':[0.001,0.005,0.01,0.05,0.1,0.5,0.99]
    },
    'lasso':{
        'lasso__alpha':[0.001,0.005,0.01,0.05,0.1,0.5,0.99]
    },
    'enet':{
        'elasticnet__alpha':[0.001,0.005,0.01,0.05,0.1,0.5,0.99]
    },
    'reg':{
        'linearregression__fit_intercept':[True],
        'linearregression__normalize':[False],
        'linearregression__n_jobs':[True]
    }
}
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.exceptions import NotFittedError
```

```python
fit_models={}
for algo, pipeline in pipelines.items():
    model =GridSearchCV(pipeline, hypergrid[algo], cv=10, n_jobs=-1)
    try:
        print(f"Starting training for {algo} has been fit")
        model.fit(X_train,y_train)
        fit_models[algo]=model
        print(f"{algo} has been fit")
    except NotFittedError as e:
        print(repr(e))
```

```python
from sklearn.metrics import r2_score, mean_absolute_error
```

```python
r2scores = []
maescores = []
rmsescores = []
for algo, model in fit_models.items():
    yhat = model.predict(X_test)
    r2scores.append(r2_score(y_test, yhat))
    maescores.append(mean_absolute_error(y_test, yhat))
    rmsescores.append(np.sqrt(mean_absolute_error(y_test, yhat)))
    print(f'{algo} scores - R2: {r2_score(y_test, yhat)} MAE: {mean_absolute_error(y_test, yhat)} RMSE: {np.sqrt(mean_absolute_error(y_test, yhat))}')
```

```python
algos = ['rf','gb', 'ridge', 'lasso', 'enet', 'reg']
```

```python
performance = pd.DataFrame(algos, columns = ["Algorithms"])
performance['r2scores'],performance['maescores'],performance['rmsescores'] = r2scores, maescores,rmsescores
print(performance)
performance.to_csv("vgperformances.csv")
```

Observations:

- Models perform similarly 
- Metrics give no clear signs of overfitting
- Gradient Booster and Random Forest are computationally heavy

Following Occam's Razor principle, we figured deploying the linear regression would be the most convenient stratefy. 

PS: 

- (-) the model implicitly takes into account the date; less videogames were published in 1980 than in later days

+ (+) errors are manageable, the model still gives solid indication of the success (or insuccess) of a videogame
